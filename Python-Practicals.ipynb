```python
# 10.	Write a python program to compute roots of a quadratic equation.

import math

print("Quadratic Equation: ax^2 + bx + c = 0")

a = int(input("Enter a: "))
b = int(input("Enter b: "))
c = int(input("Enter c: "))

# Discriminant
d = b*b - 4*a*c

# Checking conditions
if d > 0:
    root1 = (-b + math.sqrt(d)) / (2*a)
    root2 = (-b - math.sqrt(d)) / (2*a)
    print("Two real roots are:", root1, "and", root2)

elif d == 0:
    root = -b / (2*a)
    print("Both roots are same:", root)

else:
    print("Roots are complex (not real numbers).")
```

```python
# 37.	Write a program to read a text file & count the number of lines , words and characters in it.

file = open("sample.txt", "r")

lines = file.readlines()   # Read all lines

num_lines = len(lines)      # Count lines
num_words = 0
num_chars = 0

# Counting words and characters
for line in lines:
    words = line.split()        # Split line into words
    num_words += len(words)     # Count words
    num_chars += len(line)      # Count characters

# Display results
print("Number of lines:", num_lines)
print("Number of words:", num_words)
print("Number of characters:", num_chars)

file.close()
```

```python
# 38.	Write a program that reads data from one file & writes all uppercase words to another file

# Program to read from one file and write only uppercase words to another file

# Open source file in read mode
file1 = open("input.txt", "r")

# Open destination file in write mode
file2 = open("output.txt", "w")

# Read the content from file1
content = file1.read()

# Split into words
words = content.split()

# Write only uppercase words to output file
for w in words:
    if w.isupper():            # Check if the word is all uppercase
        file2.write(w + "\n")  # Write word to new file

# Close files
file1.close()
file2.close()

print("Uppercase words have been written to output.txt")
```

```python
'''
44. Multi-Level Aggregation Using the following DataFrame:
data = { 'City': ['Delhi', 'Delhi', 'Mumbai', 'Mumbai', 'Chennai', 'Chennai'],
              'Month': ['Jan', 'Feb', 'Jan', 'Feb', 'Jan', 'Feb'],
              'Sales': [20000, 22000, 18000, 21000, 15000, 17000] }
 df = pd.DataFrame(data)
1. Group data by City and compute total and average sales using .agg(['sum', 'mean']).
2. Sort the result by average sales in descending order.
'''

import pandas as pd

data = {
    'City': ['Delhi', 'Delhi', 'Mumbai', 'Mumbai', 'Chennai', 'Chennai'],
    'Month': ['Jan', 'Feb', 'Jan', 'Feb', 'Jan', 'Feb'],
    'Sales': [20000, 22000, 18000, 21000, 15000, 17000]
}
df = pd.DataFrame(data)

result = df.groupby('City')['Sales'].agg(['sum', 'mean'])
result = result.sort_values(by='mean', ascending=False)

print(result)
```

```python
'''
45. Realistic Data Analysis (Moderate)
data = { 'Employee': ['A', 'B', 'C', 'D', 'E', 'F'],
                               'Department': ['HR', 'IT', 'IT', 'Finance', 'Finance', 'HR'],
          'Salary': [30000, 50000, 55000, 45000, 40000, 32000],
                          'Experience': [2, 5, 3, 4, 2, 6] }
df = pd.DataFrame(data)
 Perform the following:
1.	Display the highest salary in each department.
2.	Find the average experience by department.
3.	Add a new column Bonus = 0.1 * Salary only for employees with Experience > 3.
4.	Show the department with the maximum total salary.
'''

import pandas as pd

data = {
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'Finance', 'HR'],
    'Salary': [30000, 50000, 55000, 45000, 40000, 32000],
    'Experience': [2, 5, 3, 4, 2, 6]
}

df = pd.DataFrame(data)

# 1. Highest salary in each department
highest_salary = df.groupby('Department')['Salary'].max()

# 2. Average experience by department
avg_experience = df.groupby('Department')['Experience'].mean()

# 3. Add Bonus column
df['Bonus'] = 0
# Assign bonus only where experience > 3
df.loc[df['Experience'] > 3, 'Bonus'] = df['Salary'] * 0.1

# 4. Department with maximum total salary
total_salary = df.groupby('Department')['Salary'].sum()
department_max_salary = total_salary.idxmax()

print("Highest Salary per Department:", highest_salary)
print("Average Experience per Department:", avg_experience)
print("DataFrame with Bonus Column:", df)
print("Department with Maximum Total Salary:", department_max_salary)
```

```python
# 54. Create a heatmap using Seaborn to show the correlation between columns in a DataFrame

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = { 'Math': [85, 78, 92, 88, 76],
    	  'Science': [80, 75, 89, 90, 70],
             'English': [78, 82, 85, 80, 74],
             'History': [72, 70, 78, 75, 68]}
df = pd.DataFrame(data)
corr = df.corr()

plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")

plt.show()
```

```python
# 59. Visualize a 3D surface or scatter plot using matplotlib.pyplot.axes(projection='3d')

#Scatter Plot:

import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

# Create 3D axes using plt.axes()
ax = plt.axes(projection='3d')
ax.scatter(x, y, z)

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("3D Scatter Plot")
plt.show()
```

```python
# Surface plot

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 40)
y = np.linspace(-5, 5, 40)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("3D Surface Plot")
plt.show()
```

```python
# 60. Combine Pandas and Matplotlib to plot a bar chart of total sales per region from a dataset.

import pandas as pd
import matplotlib.pyplot as plt

data = { 'Region': [ 'North','South','East','West','North','South','East','West',
                             'North','South','East','West','Central','Central','North','South'],
             'Sales': [ 20000,15000,18000,22000,25000,17000,19000,21000,
                           30000,16000,20000,23000,18000,24000,27000,15500]}

df = pd.DataFrame(data)
total_sales = df.groupby('Region')['Sales'].sum()

plt.bar(total_sales.index, total_sales.values)

plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.title("Total Sales per Region")
plt.show()
```

```python
# 62. Create an advanced visualization — for example, overlay a line plot on a bar plot using dual y-axes (twinx()).

import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [20000, 23000, 25000, 22000, 27000, 30000]
profit = [4000, 4500, 4700, 4200, 5200, 6000]

fig, ax1 = plt.subplots()

# Bar Plot (Left Y-axis)
ax1.bar(months, sales)
ax1.set_xlabel("Months")
ax1.set_ylabel("Sales", color='black')
ax1.tick_params(axis='y', labelcolor='black')

# Create twin axis
ax2 = ax1.twinx()

# Line Plot (Right Y-axis)
ax2.plot(months, profit, marker='o', color='orange')
ax2.set_ylabel("Profit", color='black')
ax2.tick_params(axis='y', labelcolor='black')

plt.title("Sales vs Profit (Bar + Line Overlay Using twinx())")
plt.show()
```
